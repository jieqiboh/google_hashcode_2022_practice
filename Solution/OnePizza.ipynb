```python
fileNames = ["a_an_example", "b_basic", "c_coarse", "d_difficult", "e_elaborate"]  # File names
inputFilesDirectory = '../Input/'
outputFilesDirectory = '../Output/'
```

```python
def processFile(fileName,inputFilesDirectory):
    inputFile = open(inputFilesDirectory + fileName + ".in.txt", "rt")
    
    numClients = 0
    #all ingredients
    ingredients = []
    #ingredients that are liked
    liked_ingredients = []
    #ingredients that are disliked
    disliked_ingredients = []
    #ingredients that are only liked
    filtered_liked_ingredients = []
    #ingredients that are only disliked
    filtered_disliked_ingredients = []
    
    #process data
    lineOne = inputFile.readline()
    numClients = int(lineOne)
    
    for i in range(numClients*2):
        line = inputFile.readline()

        numIngredients = int(line.split()[0])
        
        if numIngredients == 0:
            continue            
        else:
            #liked
            if i%2==0:
                liked_ingredients.append(line.split()[1:])
            #disliked
            else:
                disliked_ingredients.append(line.split()[1:])

    liked_ingredients = [item for sublist in liked_ingredients for item in sublist]
    disliked_ingredients = [item for sublist in disliked_ingredients for item in sublist]
#     MAX_SLICES , NUM = list(map(int, lineOne.split()))
#     PIZZAS = list(map(int, lineTwo.split()))
    return liked_ingredients,disliked_ingredients
```

```python
def solve(liked_ingredients,disliked_ingredients):
    unique_liked_ingredients = list(set(liked_ingredients))
    unique_disliked_ingredients = list(set(disliked_ingredients))
    
    #generate dictionary of unique ingredients with initial value 0
    unique_ingredients = list(set(liked_ingredients+disliked_ingredients))
    dict_unique_ingredients = {}
    for ingredient in unique_ingredients:
        dict_unique_ingredients.update({ingredient: 0})
    
    #if key appears in liked_ingredients, add 1
    #if key appears in liked_ingredients, minus 1
    for i in liked_ingredients:
        dict_unique_ingredients[i]+=1
    for i in disliked_ingredients:
        dict_unique_ingredients[i]-=1
        
    final_ingredients = []
    
    for key in dict_unique_ingredients:
        if dict_unique_ingredients[key] > 0:
            final_ingredients.append(key)
    
    return final_ingredients
```

```python
def outputFile(final_ingredients, fileName, outputFilesDirectory):    
    
    #Generate output for 1 file
    outputString = str(len(final_ingredients)) + " "
    for ingredient in final_ingredients:
        outputString = outputString + ingredient + " "
    print(outputString)

    outputFile = open(outputFilesDirectory + fileName + ".out", "w")
    outputFile.write(outputString)
    outputFile.close()
```

```python
a_an_example = processFile("a_an_example",inputFilesDirectory)
solved_a_an_example = solve(a_an_example[0],a_an_example[1])
outputFile(solved_a_an_example, "a_an_example", outputFilesDirectory)
```
